```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score 
```

```python
car_data=pd.read_csv('car data.csv')
```

```python
car_data.head()
```

```python
car_data.replace({'Fuel_Type':{'Diesel':1,'Petrol':0,'CNG':2}},inplace=True)
car_data.replace({'Seller_Type':{'Dealer':1,'Individual':0}},inplace=True)
car_data.replace({'Transmission':{'Manual':1,'Automatic':0}},inplace=True)
car_data = car_data.infer_objects(copy=False)
```

```python
car_data.head()
```

```python
X=car_data.drop(['Car_Name','Selling_Price'],axis=1)
Y=car_data['Selling_Price']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=2)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
training_data=model.predict(X_train)
```

```python
result_train=r2_score(Y_train,training_data)
print(result_train)
```

```python
test_data=model.predict(X_test)
```

```python
result_test=r2_score(Y_test,test_data)
print(result_test)
```

```python
plt.scatter(Y_train,training_data, c='blue', label='X vs Y')  # Blue points

plt.legend()
plt.show()
```

```python
#through lasso regression
model1=Lasso()
```

```python
model1.fit(X_train,Y_train)
```

```python
lasso_train_data=model1.predict(X_train)
lasso_train_result=r2_score(Y_train,lasso_train_data)
print(lasso_train_result)
```

```python
lasso_test_data=model1.predict(X_test)
lasso_test_result=r2_score(Y_test,lasso_test_data)
print(lasso_test_result)
```
